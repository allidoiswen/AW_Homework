```python
import pandas as pd
import seaborn as sns
from sqlalchemy import create_engine
from config import db_key
```

```python
# 1. Import the SQL database into Pandas. (Yes, you could read the CSVs directly in Pandas, 
# but you are, after all, trying to prove your technical mettle.) This step may require some research.
engine = create_engine(f'postgresql://{db_key}@localhost:5432/EmployeeSQL')
connection = engine.connect()
```

```python
# 2. Create a histogram to visualize the most common salary ranges for employees.
salary_df = pd.read_sql("SELECT salary FROM salary;", connection)

ax = sns.distplot(salary_df['salary'], kde = False,
             hist_kws = {"histtype": "step", "linewidth": 4, "alpha": 1, "color": "g"}, 
             bins = 10,
             axlabel = 'Salary').set_title('Employees Salary Distribution')
ax
```

```python
# 3. Create a bar chart of average salary by title.
ave_salary_df = pd.read_sql('SELECT title, AVG(s.salary) AS "Average Salary" \
                             FROM employees AS e \
                             INNER JOIN salary AS s ON e.emp_no = s.emp_no \
                             INNER JOIN title AS t ON e.emp_title = t.title_id \
                             GROUP BY title;'
                            ,connection)

sns.set(style = "whitegrid")
ax = sns.barplot(x = ave_salary_df['title'], y = ave_salary_df['Average Salary'])
ax.set_xticklabels(ave_salary_df['title'], rotation = 70)
ax.set_title('Average Salary by Title');
```
